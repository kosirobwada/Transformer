```python
text = """I am extremely disappointed with the ABC Electric Kettle. After reading some positive reviews, I decided to give it a try, but it has been nothing but trouble. The first issue I encountered was with the build quality. The plastic feels cheap and flimsy, and the lid doesn’t close properly, which causes water to spill out when pouring.The kettle is supposed to boil water quickly, but it takes much longer than expected. It’s advertised to boil a full kettle in under five minutes, but mine takes nearly ten minutes every time. Additionally, it makes a loud buzzing noise during operation, which is very annoying.To make matters worse, the kettle stopped working entirely after just three weeks of use. It refuses to turn on, even though I have checked the power outlet and tried different sockets. I contacted customer service, but their response was slow, and they were unhelpful. They offered to replace it, but I have little confidence that the new unit will be any better.Overall, the ABC Electric Kettle has been a huge disappointment. Poor build quality, slow performance, and terrible customer service make this a product I cannot recommend to anyone. Save your money and look for a more reliable option."""
```

```python
from transformers import pipeline
classifier = pipeline("text-classification")
```

```python
import pandas as pd

outputs = classifier(text)
pd.DataFrame(outputs)
```

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple")
outputs = ner_tagger(text)
pd.DataFrame(outputs)
```

```python
reader = pipeline("question-answering")
question = "What does the customer want?"
outputs = reader(question=question, context = text)
pd.DataFrame([outputs])
```

```python
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=45, clean_up_tokenization_spaces=True)
print(outputs[0]["summary_text"])
```

```python
translator = pipeline(
    "translation_en_to_de",
    model = "HelSinki-NLP/opus-mt-en-de"
)
```

```python
outputs = translator(text, clean_up_tokenization_spaces=True, min_length=100)
```

```python
print(outputs[0]['translation_text'])
```

```python
generator = pipeline("text-generation")
response = "Dear Bumblee, I am sorry to hear that your order was mixed up!"
prompt = text + "\n\nCustomer service response:\n" + response
outputs = generator(prompt, max_length=200)
print(outputs[0]['generated_text'])
```
